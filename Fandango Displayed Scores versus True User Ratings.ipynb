```python
df = pd.merge(fandango,all_sites,on='FILM',how='inner')
```

```python
df.info()
```

```python
df.head()
```

```python
df['RT_Norm']=np.round(df['RottenTomatoes']/20,1)
df['RTU_Norm']=np.round(df['RottenTomatoes_User']/20,1)
```

```python
df['Metacritic_Norm']=np.round(df['Metacritic']/20,1)
df['MTU_Norm']=np.round(df['Metacritic_User']/2,1)
```

```python
df['IMDB_Norm']=np.round(df['IMDB']/2,1)
```

```python
df.head()
```

```python
norm_scores=df[['STARS','RATING','RT_Norm','RTU_Norm','Metacritic_Norm','MTU_Norm','IMDB_Norm']]
```

```python
norm_scores.head()
```

```python
def move_legend(ax, new_loc, **kws):
    old_legend = ax.legend_
    handles = old_legend.legend_handles
    labels = [t.get_text() for t in old_legend.get_texts()]
    title = old_legend.get_title().get_text()
    ax.legend(handles, labels, loc=new_loc, title=title, **kws)
```

```python
fig,ax=plt.subplots(figsize=(15,6),dpi=200)
sns.kdeplot(data=norm_scores,fill=True,clip=[0,5],ax=ax,palette='Set1')
move_legend(ax,'upper left')
```

```python
fig, ax = plt.subplots(figsize=(15,6),dpi=150)
sns.kdeplot(data=norm_scores[['RT_Norm','STARS']],clip=[0,5],fill=True,palette='Set1',ax=ax)
move_legend(ax, "upper left")
```

```python
plt.subplots(figsize=(15,6),dpi=150)
sns.histplot(data=norm_scores,bins=50)
```

```python
sns.clustermap(norm_scores,col_cluster=False,cmap='magma')
```

```python
norm_Col=df[['STARS','RATING','RT_Norm','Metacritic_Norm','MTU_Norm','IMDB_Norm','FILM']]
norm_Col.nsmallest(10,'RT_Norm')
```

```python
plt.figure(figsize=(10,4),dpi=200)
worst_films=norm_Col.nsmallest(10,'RT_Norm').drop('FILM',axis=1)
sns.kdeplot(data=worst_films,fill=True,clip=[0,5])
plt.title("Ratings for RT Critic's 10 Worst Reviewed Films")
plt.show()
```
